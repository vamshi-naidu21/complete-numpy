*View or Shallow Copy


NumPy has a view() method which is a new array object that looks at the same 

data of the original array. In this case, change in dimensions of the new array doesn't change dimensions of the original.**

```
# declare an array

x1 = np.arange(6).reshape(3,2) 

print(x1) 
```

```
# create view of x1

y1 = x1.view()

print(y1)
```

```
y1[0,0] = 100 

print('Modified array y1:')

print(y1)  
```

*Deep Copy


The copy() function creates a deep copy. It is a complete copy of the array and its data, and doesn't share with the original array.**

```
# declare an array

z1 = np.array([[10,10], [2,3], [4,5]]) 

print('Array z1 is:')

print(z1)  
```

```
# create a deep copy of z1

y1 = z1.copy()

print('Array y1 is:')

print(y1) 
```

```
# y1 does not share any memory of x1

print('Can we write y1 is z1')

print(y1 is z1)  
```

```
# change the contents of y1

y1[0,0] = 100 

print('Modified array y1:')

print(y1)  
```

```
# z1 remains unchanged

print(z1)
```
